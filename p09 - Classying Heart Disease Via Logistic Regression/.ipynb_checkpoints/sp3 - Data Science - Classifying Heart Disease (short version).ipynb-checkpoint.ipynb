## Data Set Information
---

[From the source](https://archive.ics.uci.edu/ml/datasets/Heart+Disease):


This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them. In particular, the Cleveland database is the only one that has been used by ML researchers to

this date. The "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4. Experiments with the Cleveland database have concentrated on simply attempting to distinguish presence (values 1,2,3,4) from absence (value 0).

Attribute Information:

- `age` - age in years.


- `sex`:
    - 1 = male
    - 0 = female


- `cp`, chest pain type:
    - `Value 1`: typical angina.
    - `Value 2`: atypical angina.
    - `Value 3`: non-anginal pain.
    - `Value 4`: asymptomatic.
    
    
- `trestbps` - resting blood pressure (in mm Hg (milliliters of mercury) on admission to the hospital).



- `chol` - serum cholesterol in mg/dl (levels in milligrams per deciliter).


- `fbs` - (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) - test measures sugar (glucose) in the blood:
    - 99 mg/dL or lower: This is a normal fasting blood sugar level.
    - 100â€“125 mg/dL: Fasting blood sugar in this range typically indicates prediabetes. This means your blood sugar levels are higher than normal but not high enough to be classified as diabetes.
    - 126 mg/dL or above: This indicates high blood sugar, the main sign of diabetes.


- `restecg` - resting electrocardiographic results:
    - `Value 0`: normal.
    - `Value 1`: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV).
    - `Value 2`: showing probable or definite left ventricular hypertrophy by Estes' criteria.
    
    
    
- `thalach` - maximum heart rate achieved.


- `exang` - exercise induced angina: 
    - 1 = yes.
    - 0 = no.
    
    
- `oldpeak` = [ST depression](https://en.wikipedia.org/wiki/ST_depression) induced by exercise relative to rest (mm).


- `slope` - the slope of the peak exercise ST segment:
    - `Value 1`: upsloping.
    - `Value 2`: flat.
    - `Value 3`: downsloping.
    
    
    
- `ca`: number of major vessels (0-3) colored by fluoroscopy.



- `thal` - presence of thalassemia (a blood disorder): 
    - 3 = normal.
    - 6 = fixed defect (no blood flow in some part of the heart).
    - 7 = reversible defect (a blood flow is observed but it is not normal).
    
    
    
- `num`/`present` - diagnosis of heart disease (angiographic disease status)
    - `Value 0`: < 50% diameter narrowing.
    - `Value 1`: > 50% diameter narrowing.

### Preliminary steps:
- drop `ca`
- transform into dummies `cp` (chest pain), `restecg` (resting electrocardiographic results), `thal` (presence of thalassemia) and `slope` (ST segment slope).
- analyzing boxplots of predictors controlling for  the target value.


```python
heart =  heart.drop(columns='ca')

cp_dict = {
            1: 'typical_angina',
            2: 'atypical_angina',
            3: 'non-anginal_pain',
            4: 'asymptomatic',
            }

restecg_dict = {
            0: 'normal',
            1: 't_wave_abnormality',
            2: 'ventricular_hypertrophy',
            }

thal_dict = {
            '?': 'inconclusive',
            '3.0': 'normal',
            '6.0': 'fixed_defect',
            '7.0': 'reversible_defect',
            }

slope_dict = {
            1: 'upsloping',
            2: 'flat',
            3: 'downsloping',
            }

heart.cp = heart.cp.map(cp_dict)

heart.restecg = heart.restecg.map(restecg_dict)

heart.thal = heart.thal.map(thal_dict)

heart.slope= heart.slope.map(slope_dict)
```

Before we turn the categorical variables into dummies we look at their frequency distribution in order to exclude rows for which the value is too irrelevant. 

```python
numerical = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak',]
categorical = [col for col in heart.columns if col not in numerical]

for col in categorical:
    print(f'Variable: {col}')
    print(heart[col].value_counts())
    print('\n')
```

As we can observe in the values above, two variables present very uneven numbers: `restecg` has only 4 rows with `t_wave_abnormality` and `thal` has only 2 rows for `inconclusive`. All other variables are fairly well distributed.

We proceed to drop the aforementioned rows and turn all the categorical variables into dummies.

```python
to_drop = heart.query('restecg == "t_wave_abnormality" | thal == "inconclusive"')

heart = heart.drop(index=to_drop.index)

heart = pd.get_dummies(data=heart,
                       columns=['cp', 'restecg', 'thal', 'slope']
                      ).drop(columns=['restecg_normal', 'cp_asymptomatic', 'thal_normal', 'slope_upsloping'])
```

We show the basic information of `heart` again in order to assess the modifications made so far.

```python
heart.info()
```

### Removing outliers.

In order to achieve more accurate predictions we have to assure ourselves that the coefficients aren't biased due to influential observations. To prevent that, we'll be trimming values for non-binary variables that exceed 2 standard deviations. 

```python
all_feats = [col for col in heart.columns if col != 'present']

non_binary = [col for col in heart.columns if len(heart[col].unique()) != 2]
      
non_binary
```

Dropping rows that contain a value with a z-score over 2.

```python
from scipy.stats import zscore

z_scores = zscore(heart[non_binary])

below_2_sdev = z_scores[z_scores < 2].dropna(axis=0)

heart = heart.loc[below_2_sdev.index ,:].reset_index(drop=True)

f'Rows left after trimming outliers: {heart.shape[0]}'
```

### Variance Inflation Factor(VIF) analysis.

As we'll be able to see below, the model with all the features shows that, when we do a VIF analysis, there is evidence of high multicollinearity among some of the predictors: all values above 4 must be a cause of concern. To circumvent this problem we withdraw the variables with higher VIF and run again the auxiliary regressions; the new VIF results show that withdrawing those variables, in general, help to reduce the highest VIF values. We'll be using this models to compute logistic regressions in order to compare their predictions.   

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

features = [col for col in heart.columns if col != 'present']

feats_dict = {
    'all_features': features,
    'no_age': [col for col in features if col not in ['age']],
    'no_thalach': [col for col in features if col not in ['thalach']],
    'no_age_thalach': [col for col in features if col not in ['thalach', 'age']]
             }


# VIF DataFrame.
vif_data = pd.DataFrame(index=heart[features].columns)
  
for feat_set in feats_dict:
    feats = feats_dict[feat_set]
    vif_list = [vif(heart[feats].values, i) for i in range(len(heart[feats].columns))]
    vif_series = pd.Series(vif_list, index=feats, name=feat_set)
    
    vif_data = pd.concat([vif_data, vif_series], axis=1)

vif_data
```

## Model Fitting And Prediction Making
---
For this analysis we randomly split the train and the test set into 60 and 40% sub-sample sizes, respectively.

```python
from sklearn.model_selection import train_test_split as split

X = heart[features]

y = heart.present

X_train, X_test, y_train, y_test = split(X,
                                         y,
                                         test_size=0.4,
                                         random_state=342)
```

### Part I - Using the train set to fit and make "control" predictions.

Checking the distribution of `present` in the train and test sets.

```python
pd.DataFrame(
    {
        'train': y_train.value_counts(),
        'test': y_test.value_counts()
    })
```

As we can see, the values of the target variable are fairly evenly distributed.

In the part, we not only fit and produce predictions resorting solely to the train set, we calculate the prediction metrics that allow us to assess the predicting qualities of the model, these are: 

- accuracy (overall capability to predict the outcome).
- sensitivity (capability to successfully predict cases of heart disease).
- specificity (capability to successfully predict cases of absence of heart disease).

#### How does the process works?

The function `log_model()` takes as input a list of variable names to be fitted as independent variables, an `X` DataFrame and a target Series, `y` and returns an object that contains the fitted model.

`reurn_scores()`, the same style of inputs as the previous function but also takes the fitted model return by `log_model()` in order to make predictions, calculate and return the three aforementioned prediciton metrics. The input `y`, in this case, serves to calculate those metrics.

```python
from sklearn.linear_model import LogisticRegression

def log_model(features, X, y):
    logistic = LogisticRegression(max_iter=10000)
    model = logistic.fit(X[features].values, y.values)
    return model

def return_scores(features, model, X, y):
    prediction = model.predict(X[features].values)
    
    # Accuracy.
    accuracy = sum(prediction == y) / y.shape[0]
    accuracy = round(accuracy*100, 2)
    
    # Sensitivity.
    true_positives =  sum((prediction == y) & (prediction == 1))
    false_negatives = sum((prediction != y) & (prediction == 0))
    
    sensitivity = true_positives / (true_positives + false_negatives)
    sensitivity = round(sensitivity*100, 2)
    
    # Specificity.
    true_negatives = sum((prediction == y) & (prediction == 0))
    false_positives = sum((prediction != y) & (prediction == 1))
    
    specificity = true_negatives / (true_negatives + false_positives)
    specificity = round(specificity*100, 2)
    
    results = {
        'accuracy': accuracy,
        'sensitivity': sensitivity,
        'specificity': specificity
    }
    
    return results
    
```

```python
scores_train_df = pd.DataFrame()

models_dict = {}

for feat_set in feats_dict:
    
    feats = feats_dict[feat_set]
    
    model = log_model(feats, X_train, y_train)
    
    models_dict[feat_set] = model
    
    scores_train = return_scores(feats, model, X_train, y_train)
    
    scores_train_df = pd.concat(
        [scores_train_df, pd.Series(scores_train, name=feat_set)],
        axis=1)
    
scores_train_df
```

As we can see above, all models behave similarly while the model that with all features performs a little better than the rest in terms of accuracy and specificity (85.99% and 90.62% respectively) but the sensitivity is the same for the models with all features, no age and no 'thalach' (maximum heart rate achieved): 78.69%. The model that performed better in terms of the VIF analysis performed slightly worst in all metrics than all other models. Moreover, these values, in terms of accuracy, are not conclusive in determining whether the model is prone to overfitting or not.


#### Analyzing the model without `age` and `thalahc` - `no_age_thalach`. 


```python
log_odds_coefs = models_dict['no_age_thalach'].coef_[0]

odds_ratio = np.exp(log_odds_coefs)

odds_ratio = pd.Series(
    odds_ratio, 
    index=feats_dict['no_age_thalach'],
    name='odds ratio coeffs'
    )

log_odds = pd.Series(
    log_odds_coefs,
    index=feats_dict['no_age_thalach'],
    name='log odds coeffs'
    )

odds_ratio_percentage = round((odds_ratio - 1)*100, 2)
odds_ratio_percentage.name = 'odds ratio percentage'


coeffs = pd.concat(
    [log_odds, odds_ratio, odds_ratio_percentage],
    axis=1
    )

coeffs = coeffs.sort_values('odds ratio coeffs', ascending=False)

coeffs
```

The data frame `coeffs` gives us the log odds and odds ratios for the `no_thalach` model. The order given is based on the odds ratio (second column) in descending order. Looking to the odds ratio column we see that only 7 variables are above 1, i.e. when present, they increase the chances of heart disease:



- The dummies related with the presence of a blood disorder called thalassemia increase 3.37 and 2.06 times (237 and 106 in percentage), respectively, the chances of developing heart disease. These are the most impactfull factors of those considered.


- Next, being male, `sex=1`, increase the chances of heart disease by 151.62%.


 - ST segment slope: down slopping or a flat response of the ST segment during peak exercise (`slope_downsloping` and `slope_flat` respectively) [may indicate heart disease](https://ecgwaves.com/st-segment-normal-abnormal-depression-elevation-causes/) but in this model only the latter indicates that, where the presence of a flat ST segment increases the chance of heart disease by 150.9%.


- The dummies related with resting electrocardiographic results (`estecg`) show that if there is evidence of probable or definite left ventricular hypertrophy - `restecg_ventricular_hypertrophy` the chances of developing a heart disease increases 110%.



- Still on the analysis of the ST segment, the increse of the ST depression induced by exercise relative to rest (in mm) - `oldpeak`, increases the chances of heart disease by 66.39%.



- Lastly, an exercise induced angina `exang=1` increase the chances of heart disease by 62.49%.


Regarding variables that, according to the model, decrease the risk of heart disease, we may dismiss them, since in theory, they may or may not contribute to heart disease, but they are very unlikely to actually decrease the chances of heart disease. In the case of `fbs` there is some [evidence](https://publichealth.jhu.edu/2005/selvin-bloodsugar) that high blood sugar can contribute independently to increasing heart disease risk, thus contradicting the findings. Moreover, variables related to chest pain type (`cp`) - `cp_typical_angina`, `cp_non-anginal_pain`, `cp_atypical_angina`, are expected to have some relation with heart disease, [specially angina related pain](https://www.mayoclinic.org/diseases-conditions/angina/symptoms-causes/syc-20369373), but the model does not capture that relation.

 

### Part II - calculating predictions based on the test set and compare the prediction metrics.

```python
scores_test_df = pd.DataFrame()

models_dict = {}

for feat_set in feats_dict:
    
    feats = feats_dict[feat_set]
    
    model = log_model(feats, X_train, y_train)
    
    models_dict[feat_set] = model
    
    scores_test = return_scores(feats, model, X_test, y_test)
    
    scores_test_df = pd.concat(
        [scores_test_df, pd.Series(scores_test, name=feat_set)],
        axis=1)
    
scores_test_df
```

The following table, `compare`, displays the percentage difference obtained by comparing the test set metrics with the train set metrics.

```python
compare = (scores_test_df - scores_train_df) / scores_train_df

compare = compare.round(3)*100

compare
```

From both `scores_test_df` and `compare` (scores) tables, we can see that, in general, the accuracy decreased from the train set prediction around 5 to 6%. Sensitivity increased substantially for the model with all variables whilst roughly maintaining the same values for the other models. In terms of specificity, the best model is now the one without `age` and `thalach`, which saw the smallest reduction in that metric: -5.9%. In regards to the best model, no version has the best values all around. The model with all variables performs better in terms of sensitivity, which is arguably the most important metric, since the we are more interest in predicting cases (heart disease) than predicting non-cases (no heart disease). Conversely, the model chosen as reference - the model with no `age` and no `thalach`, performs substantially worst than the largest model by almost 10%, when comparing the test set sensitivity metric - 84.62 and 75%, respectively. Moreover, the former model actually performs much better than the latter model (all variables) in terms of specificity by a difference of 7.4%.
