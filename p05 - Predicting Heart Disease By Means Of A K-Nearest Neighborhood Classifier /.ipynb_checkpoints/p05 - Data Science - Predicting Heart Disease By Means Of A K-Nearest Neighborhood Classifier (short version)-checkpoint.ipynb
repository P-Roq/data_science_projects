# Predicting Heart Disease By Means Of A K-Nearest Neighborhood Classifier

## Introduction
---
Given a [data set](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction) with medical information on patients that have been subject to cardiovascular tests, we are tasked to design and train a k-nearest neighborhood (KNN) classifier that successfully predicts whether a patient has developed a heart complication or not.

Data set description:

|Feature | Description            |
|-----|---------------------------|
|`Age`  | Age of the patient in years.|
|`Sex`  | Sex of the patient (M: Male, F: Female).|
|`ChestPainType`  | Chest pain type (TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: |Asymptomatic). |
|`RestingBP`  | Resting blood pressure (mm Hg).|
|`Cholesterol`  | Serum cholesterol (mm/dl).|
|`FastingBS`  | Fasting blood sugar (1: if FastingBS > 120 mg/dl, 0: otherwise).|
|`RestingECG`  | Resting electrocardiogram results (Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria).|
|`MaxHR`  | Maximum heart rate achieved (Numeric value between 60 and 202).|
|`ExerciseAngina`  | Exercise-induced angina (Y: Yes, N: No).|
|`Oldpeak`  | ST numeric value measured in depression in mm).|
|`ST_Slope`  | The slope of the peak exercise ST segment (Up: 'upsloping', Flat: 'flat', Down: 'downsloping').|
|`HeartDisease`  | Output class (1: heart disease, 0: Normal).|

### Replacing invalid values in `cholesterol` with a sub-sample median.

Since `cholesterol` has 172 invalid observations, we chose to replace those values with a median taken out of a sub-sample that controls for features that may affect the level of cholesterol, in this case: `age`, `sex` and `restingecg`. For `age` instead of using this variable as is, we are going to use a dummy version that simply distinguishes from patients under 50 years old and older patients. `sex` has also two subcategories and `restingecg` has three - `Normal`, `ST` and `LVH`, therefore, the sub-samples we want to extract the medians from, are combinations of those sub-categories. `medians_support_dict` contains all the combinations of these three features. The process to find the appropriate medians and replace the `cholesterol` invalid values with them is the following:


1. Create the dummy variable for `age` in order to distinguish people with less than 50 years old from the rest.


2. Create a dictionary - `medians_support_dict`, that will help to create sub-samples in `heart` with the combinations of `age_dummy`, `sex` and `restingecg`.


3. Create a function - `get_median()`, that takes the values of `medians_support_dict`, which are lists with combinations of the sub-categories of the aforementioned features; fetches the correct sub-sample; and returns the associated median for that sub-sample. 



4. Apply the function to the items in `medians_support_dict` in order to store the resulting medians in `medians_dict`. In both of these dictionaries each value/median is correctly associated with a 'code' string as key (same keys as in `medians_support_dict`) that are the combinations of the sub-categories of the aforementioned features: example `'50_and_over_M_Normal'` is a string that identifies the median for the sub-sample of observations that are 50 years old and older, male and have a normal resting ECG. 


5. Create a support series - `median_support_series`, that we'll use to map `medians_dict` in the next step. This series is build by adding the values of, in the following order: `age_dummy`, `sex` and `restingecg` (as strings); each element being separated by an underscore, "_".  


6. We apply the dictionary `medians_dict` in `median_support_series` to produce a series - `heart.cholesterol_medians` that maps correctly the sub-sample medians by their correct combination of features.  


7. Inside the `heart` DataFrame, we clone `cholesterol` into a new column - `cholesterol_2`; collect the index of the rows in `cholesterol` that are invalid (`0`); and use it to replace the same invalid values in `cholesterol_2` with the correspondent values (same index) in `heart.cholesterol_medians`.

```python
# 1.
heart["age_dummy"] = heart.age.apply(lambda x: "50_and_over" if x > 49 else "under_50")

# 2.
medians_support_dict = {
    '50_and_over_M_Normal': ['50_and_over', 'M', 'Normal'],
    '50_and_over_M_ST': ['50_and_over', 'M', 'ST'],
    '50_and_over_M_LVH': ['50_and_over', 'M', 'LVH'],
    'under_50_M_Normal': ['under_50', 'M', 'Normal'],
    'under_50_M_ST': ['under_50', 'M', 'ST'],
    'under_50_M_LVH': ['under_50', 'M', 'LVH'],
    '50_and_over_F_Normal': ['50_and_over', 'F', 'Normal'],
    '50_and_over_F_ST': ['50_and_over', 'F', 'ST'],
    '50_and_over_F_LVH': ['50_and_over', 'F', 'LVH'],
    'under_50_F_Normal': ['under_50', 'F', 'Normal'],
    'under_50_F_ST': ['under_50', 'F', 'ST'],
    'under_50_F_LVH': ['under_50', 'F', 'LVH'],
}

# 3.
def get_median(l):
    age = l[0]
    sex = l[1]
    ecg = l[2]
    
    median = heart.query("age_dummy == @age & sex == @sex & restingecg == @ecg")['cholesterol'].median()
        
    return median
    
# 4.
medians_dict = {key: get_median(medians_support_dict[key]) for key in medians_support_dict}

medians_dict
```

```python
# 5.
median_support_series = heart.age_dummy.astype(str) + '_' + heart.sex + '_' + heart.restingecg

median_support_series.head()
```

```python
# 6.
heart["cholesterol_medians"] = median_support_series.map(medians_dict)

heart["cholesterol_medians"].head()
```

```python
# 7.
index_cholest_zero = heart.query('cholesterol == 0').index

heart['cholesterol_2'] = heart['cholesterol']

heart.loc[index_cholest_zero, 'cholesterol_2'] = heart.loc[index_cholest_zero, 'cholesterol_medians']
```

As we can see below, this slice of `heart` shows that the invalid values in `cholesterol` were successfully replaced by values in `cholesterol_medians`. We can check it by comparing the medians in `cholesterol_2` with the medians in `medians_dict` via the keys.

```python
heart.query('cholesterol == 0')[['age', 'sex', 'restingecg', 'cholesterol', 'cholesterol_medians', 'cholesterol_2']].head()
```

## Model Estimation
---
The data is to be split into train (60%), validation (20%) and test (20%) sets.


### Calculating accuracy scores for the validation set.

The function below, `obtain_score()`, takes as input: 
- training and testing sets;
- the features' names selected to serve as independent variables;
- the number of neighbors in the model;
- and the type of score to return: 
    - `'accuracy'`, it measures how successful the model is in predicting any outcome (in our context, cases and non-cases of heart disease).
    - `'sensitivity'`, also designated as recall, it measures how successful the model is in predicting cases oh heart disease.
    - `'specificity'`, it measures how successful the model is in predicting non-cases oh heart disease.

The number of neighbors will be inserted as part of a dictionary of `KNeighborsClassifier` parameters which include the weights attributed to each point and which frame is used to calculate the distances between points. For now we'll leave them at their default values: `weights='uniform'` and `p=2` (euclidean distance), changing only the number of neighbors.  

```python
def obtain_score(features, X_train, y_train, X_test, y_test, param_dict, score):
    
    knn = KNC(
        n_neighbors=param_dict['n'],
        weights=param_dict['weights'],
        p=param_dict['p']
        )
    
    length = y_test.size

    if isinstance(features, str):
        features = [features]
        
    X_train = X_train[features].values
    X_test = X_test[features].values
    y_train = y_train.values
    y_test = y_test.values
    
    model = knn.fit(X_train, y_train)
    
    prediction = model.predict(X_test)
    
    if score == 'accuracy':
        accuracy = sum(y_test == prediction) / length
        return round(accuracy*100, 2)
    
    if score == 'sensitivity':
        true_pos = sum((y_test == 1) & (y_test == prediction))
        all_pos = y_test[y_test == 1].size 
        sensitivity = true_pos / all_pos  
        return round(sensitivity*100, 2) 
    
    if score == 'specificity':
        true_neg = sum((y_test == 0) & (y_test == prediction))
        all_neg = y_test[y_test == 0].size 
        specificity = true_neg / all_neg  
        return round(specificity*100, 2)       
```

### Displaying the KNN classifier accuracy and sensitivity scores for univariate models.

In this experiment, we use the validation set to produce the predicted values and calculate the scores. To that end, we build a support function - `univar_df()`, that takes two random states which allow to randomly split `heart` into different train, validation and test set versions; it also takes an error score type (as a string) so that the function may return that calculated error score for each univariate model (features as columns) and for each model version (rows), where the number of neighbors `n` is set within a range between 1 and 10. 

```python
def univar_df(rand_stat_1, rand_stat_2, score):

    feats = [col for col in abs_corr_over_thres.index if col != 'heartdisease']

    X = heart[feats]

    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)


    univar_dict = {f: [] for f in feats}

    for f in univar_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            univar_dict[f].append(
                obtain_score(f, X_train, y_train, X_val, y_val, param_dict, score)
                )

    univar_df = pd.DataFrame(univar_dict, index=range(1, 10+1))

    return univar_df
```

#### Accuracy scores for univariate models.

```python
univar_accuracy_df = univar_df(87, 323, 'accuracy')

univar_accuracy_df['best accuracy'] = univar_accuracy_df.max(axis=1) 

univar_accuracy_df.loc['idx_max', :] = univar_accuracy_df.idxmax(axis=0).astype('int32') 

univar_accuracy_df
```

`univar_accuracy_df` gives the accuracy results for univariate K-neighborhood Classifier estimations, for `n` neighbors ranging from 1 to 10. The last column - `best_accuracy`, identifies the best score along the same row (different independent variable but same `n`), and the last row - `idx_max`, identifies the model with the highest score when maintaining the independent variable but for different values of `n` (if there are repeated highest scores the `n` displayed is the smallest). The best model, when considering its accuracy, is the `st_slope_up` model, which arrives at the highest score when `n=5`: 78.14%. The second best model is `st_slope_flat` when `n=1`: 74.86%. `exerciseangina_y` appears in third place with an accuracy of 69.95% (`n=1`).

#### Sensitivity scores for univariate models.

```python
univar_sensitivity_df = univar_df(87, 323, 'sensitivity')

univar_sensitivity_df['best sensitivity'] = univar_sensitivity_df.max(axis=1) 

univar_sensitivity_df.loc['idx_max', :] = univar_sensitivity_df.idxmax(axis=0).astype('int32') 

univar_sensitivity_df
```

For sensitivity, we observe a different scenario where `oldpeak` has the best prediction power - 94.17% (`n=5`), followed by `chestpaintype_ata` - 93.20% (`n=7`), `sex_m` - 92.23% (`n=1`) and then `st_slope_up` - 84.47% (`n=5`) and `st_slope_flat` - 75.73% (`n=1`). This values contrast a lot with what we have seen for accuracy, which can be a symptom of fitting a single binary variable while also relying on a single pseudo-random distribution (random states) to shuffle the main data set into several smaller sets; therefore, a more robust way to assess these prediction scores is to calculate the mean of a large number of estimations, e.g. 100. So, lets add 99 versions of `univar_sensitivity_df` but with different random states and calculate the mean sensitivities, to assess if the values change substantially from what we have seen so far.

#### Mean sensitivity scores: 100 estimations.

```python
univar_mean_sensitivity_df = univar_df(87, 323, 'sensitivity')

for i in range(1, 100):

    univar_mean_sensitivity_df += univar_df(i, i+100, 'sensitivity')


univar_mean_sensitivity_df = univar_mean_sensitivity_df / 100

univar_mean_sensitivity_df['best sensitivity'] = univar_mean_sensitivity_df.max(axis=1) 

univar_mean_sensitivity_df.loc['idx_max', :] = univar_mean_sensitivity_df.idxmax(axis=0).astype('int32') 

univar_mean_sensitivity_df.round(2)
```

As we can observe, the values in `univar_mean_sensitivity_df` appear to be more balanced, specially when we observe the scores for `st_slope_up` and `st_slope_flat`, which no longer present null values. Regarding the best scores, `oldpeak`and `sex_m` show much lower scores when compared with the analogous values in `univar_sensitivity_df`, while the best models are, in ascending order: `st_slope_up` (81.42%, n=5), `st_slope_flat` (76.41%, n=9), `chestpaintype_ata` (75.17%, n=9), and 
`maxhr` (72.33%, n=9).	

#### Mean accuracy scores: 100 estimations.

```python
univar_mean_accuracy_df = univar_df(87, 323, 'accuracy')

for i in range(1, 100):

    univar_mean_accuracy_df += univar_df(i, i+100, 'accuracy')


univar_mean_accuracy_df = univar_mean_accuracy_df / 100

univar_mean_accuracy_df['best accuracy'] = univar_mean_accuracy_df.max(axis=1) 

univar_mean_accuracy_df.loc['idx_max', :] = univar_mean_accuracy_df.idxmax(axis=0).astype('int32') 

univar_mean_accuracy_df.round(2)
```

Compared with the previous accuracy results shown in `univar_accuracy_df`, we can assess that the general landscape is about the same, where the highest scores in `univar_mean_accuracy_df` can be found for `st_slope_up`, which arrives at a maximum accuracy of 78.80% at `n=9` - a slightly higher value than the analogous accuracy score observed in `univar_accuracy_df` (78.14%); and for `st_slope_flat`, which is still the second best accuracy predictor - its highest value also slightly increases from 74.86% (`n=1`, in `univar_accuracy_df`) to 76.42% (`n=10`). `exerciseangina_y` appears in third place, as before, with a mean accuracy of 70.68%

```python
feats = ['st_slope_up',
         'st_slope_flat',
         'exerciseangina_y',
         'oldpeak_scaled',
         'chestpaintype_ata',
         'maxhr_scaled',
         'sex_m']
```

#### Prediction scores comparison.

Similar to `univar_df()`, `multivar_df()` is a function that returns a DataFrame that compares prediction scores for a model with multiple predictors, also allowing the number of neighbors to vary from 1 to 10. 

```python
def multivar_df(rand_stat_1, rand_stat_2, feats=feats):

    X = heart[feats]
    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)

    all_feats_dict = {'accuracy': [], 'sensitivity': [], 'specificity': []}

    for key in all_feats_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            all_feats_dict[key].append(
                obtain_score(feats, X_train, y_train, X_val, y_val, param_dict, key)
                )
            
    multivar_df = pd.DataFrame(all_feats_dict, index=range(1, 10+1))

    return multivar_df
```

`mean_multivar_df` below, displays the mean value of 100 randomized model estimations, for each score mentioned.

```python
sum_multivar_df = None

for i in  np.arange(0, 200, 2):
    if sum_multivar_df is None:
        sum_multivar_df  = multivar_df(i, i+1)
    else: 
        sum_multivar_df  += multivar_df(i, i+1)

mean_multivar_df = sum_multivar_df / 100

mean_multivar_df.loc['idx_max', :] = mean_multivar_df.idxmax(axis=0).astype('int32') 

mean_multivar_df.round(2)
```

Observing `mean_multivar_df`, we can see that accuracy scores range roughly from 78 up to 83% and its highest score happens for 9 neighbors (83.15%). Sensitivity follows a similar pattern to accuracy, reaching a maximum score of 85.83% for 9 neighbors. The range of values is also roughly the same specificity, but the highest score, 87.9%, is reached earlier, for 2 neighbors.

### Grid search cross-validation of the multivariate model (hyperparameter tunning).

In this section we use the Scikit-learn `GridSearchCV` framework in order to conduct a grid search that allows to retrieve prediction scores from several model estimations that combine different values for certain hyperparameters stored in `parameters_dict`:
- `n_neighbors`: number of neighbors ranging from 1 to 20.
- `weights`: the weight attributed to each point - uniform or weighted.
- `p`: ["power parameter for the Minkowski metric"](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) ranging from 1 to 4.

The mean scores will be calculated for accuracy and sensitivity from a 5 folder cross-validation. `GridSearchCV` automatically divides X and y input into train and validation sets, therefore, we split again `X_2` and `y_2` so that 80% is used to calculate the validation based error score while the remaining 20% will be used as the test set.  

```python
X = heart[feats]
y = heart.heartdisease

X_train, X_test, y_train, y_test = split(X, 
                                         y, 
                                         test_size=0.2, 
                                         random_state=100)


parameters_dict = {
    'n_neighbors': list(range(1, 20+1)),
    'p': list(range(1, 4+1)),
    'weights': ['uniform', 'distance']
    }

knn = KNC()


grid_search_cv = GridSearchCV(estimator=knn,
                              param_grid=parameters_dict, 
                              scoring=['accuracy', 'recall'],
                              refit=False,
                              cv=5)


grid_cv = grid_search_cv.fit(X_train, y_train)

# Resume DataFrame - all hyperparameter combinations.
grid_cv_df = pd.DataFrame(grid_cv.cv_results_)

resume_cols = ['params',
               'mean_test_accuracy',
               'std_test_accuracy',
               'rank_test_accuracy',
               'mean_test_recall',
               'std_test_recall',
               'rank_test_recall']

grid_cv_df_short = grid_cv_df[resume_cols]
```

#### The best three models according to mean accuracy and mean sensitivity (recall):

```python
highest_accuracy_hyper = grid_cv_df_short.sort_values('mean_test_accuracy', ascending=False).head(3)

highest_accuracy_hyper
```

In terms of accuracy we can see that the best three models, displayed in `highest_accuracy_hyper`, have roughly the same value of 83.6%; the number of neighbors and the weights are set to 18 and 'uniform', respectively, thus increasing distance power parameter, `p`, doesn't seem to improve the accuracy: setting `p=1` (Manhattan distance) assures the best result.

```python
highest_sensitivity_hyper = grid_cv_df_short.sort_values('mean_test_recall', ascending=False).head(3)

highest_sensitivity_hyper
```

In terms of sensitivity, the best model achieves a maximum score of 86.75% when setting 19 neighbors, the Manhattan distance and uniform weights.

## Conclusion
---
Resuming the findings, the univariate KNN classifier that only includes a dummy variable that distinguishes patients that show an up sloping ST segment in their electrocardiogram, in opposition to a flat or down sloping behavior - `st_slope_up`, is almost as good as a multivariate version of the classifier at predicting cases of heart disease, with an accuracy and sensitivity around the 80% score mark. Extending the model to multiple predictors increased the same scores around 5%, which indicates us that there are other predictors that have a small but relevant positive contribution to the prediction capabilities of the model. These are:

|Feature             |Description                                               |
|--------------------|----------------------------------------------------------|
|'st_slope_flat'     | '1' if ST segment is flat, else '0'                      |
|'exerciseangina_y'  | '1' if exercise-induced angina, else '0'                 |
|'oldpeak_scaled'    | ST depression induced by exercise relative to rest (mm). |
|'chestpaintype_ata' | '1' if chest pain is derived from atypical angina (ata), else '0' |
|'maxhr_scaled'      |  maximum heart rate achieved                             |
|'sex_m'             |  '1' if male, 0 if female                                |

Because we still don't know which one of these predictors contribute the most and the ones that don't, a future task could entail performing a forward feature selection, where we add one at a time, the variables from the univariate models, from best to worst performance, to the `st_slope_up` univariate model, and assess whether we notice a relevant increase in the predictions scores.

Still on feature selection, we opted out to exclude (serum) `cholesterol`, even though we tried to modify this variable by distinguish patients with more than 50 years and cholesterol over 200 mm/dl, attempting to capture a correlation with the target variable over 30%, which we were not able to establish. 

Improvements to the model can be achieved through different methods, such as including other important features to the analysis that were not present in the 'heart_disease_prediction.csv' data set; examples of those features are the level of LDL cholesterol in the patients blood; a categorical variable that describes the smoking behavior of the patient e.g. 'non-smoker', 'light-smoker' and 'heavy smoker'; or even the body mass index (BMI) of the patient. Other methods could entail to better engineer the current features, resorting to a deeper understanding of the field of cardiology; we can also try to apply other methodologies used for the purpose of outcome prediction and compare the results with th present findings e.g. decision tree classifier, random forest classifier, logistic regression.
