## Data Cleaning

As seen previously, the data cleaning to be done is:

### Replacing invalid values in `cholesterol` with a sub-sample median.


Since `cholesterol` has 172 invalid observations we'll opt out by replacing those values with a median taken out of a sub-sample that controls for features that may affect the level of cholesterol, in this case: `age`, `sex` and `restingecg`. For `age` instead of using this variable as is, we are going to use a dummy version that simply distinguishes from patients under 50 years old and patients 50 years old and older. `sex` has also two subcategories and `restingecg` has three - `Normal`, `ST` and `LVH`, therefore the sub-samples we want to extract the medians from are combinations of those sub-categories. `medians_support_dict` contains all the combinations of these three features. The process to find the appropriate medians and replace the `cholesterol` invalid values with them is the following:


1. create the dummy variable for `age` in order to distinguish people with less than 50 years old from the rest.


2. create a dictionary - `medians_support_dict`, that will help to create sub-samples in `heart` with the combinations of `age_dummy`, `sex` and `restingecg`.


3. create a function - `get_median()`, that takes the values of `medians_support_dict`, which are lists with combinations of the sub-categories of the aforementioned features, fetches the correct sub-sample, and returns the associated median for that sub-sample. 


4. apply the function to the items in `medians_support_dict` in order to store the resulting medians in `medians_dict`. In both of these dictionaries each value/median is correctly associated with a 'code' string as key (same keys as in `medians_support_dict`) that are the combinations of the sub-categories of the aforementioned features: example `'50_and_over_M_Normal'` is a string that identifies the median for the sub-sample of observations that are 50 years old and older, male and have a normal resting ECG. 


5. create a support series - `median_support_series`, that we we'll use to map `medians_dict` in the next step. This series is build by adding the values of, in the following order, `age_dummy`, `sex` and `restingecg` (as strings), each element being separated by an underscore, "_".  


6. we apply the dictionary `medians_dict` in `median_support_series` to produce a series - `heart.cholesterol_medians` that maps correctly the sub-sample medians by their correct combination of features.  


7. inside the `heart` DataFrame, we clone `cholesterol` into a new column - `cholesterol_2`. We collect the index of the rows in `cholesterol` that are invalid (`0`), and use it to replace the same invalid values in `cholesterol_2` with the correspondent values (same index) in `heart.cholesterol_medians`.

```python
# 1.
heart["age_dummy"] = heart.age.apply(lambda x: "50_and_over" if x > 49 else "under_50")

# 2.
medians_support_dict = {
    '50_and_over_M_Normal': ['50_and_over', 'M', 'Normal'],
    '50_and_over_M_ST': ['50_and_over', 'M', 'ST'],
    '50_and_over_M_LVH': ['50_and_over', 'M', 'LVH'],
    'under_50_M_Normal': ['under_50', 'M', 'Normal'],
    'under_50_M_ST': ['under_50', 'M', 'ST'],
    'under_50_M_LVH': ['under_50', 'M', 'LVH'],
    '50_and_over_F_Normal': ['50_and_over', 'F', 'Normal'],
    '50_and_over_F_ST': ['50_and_over', 'F', 'ST'],
    '50_and_over_F_LVH': ['50_and_over', 'F', 'LVH'],
    'under_50_F_Normal': ['under_50', 'F', 'Normal'],
    'under_50_F_ST': ['under_50', 'F', 'ST'],
    'under_50_F_LVH': ['under_50', 'F', 'LVH'],
}

# 3.
def get_median(l):
    age = l[0]
    sex = l[1]
    ecg = l[2]
    
    median = heart.query("age_dummy == @age & sex == @sex & restingecg == @ecg")['cholesterol'].median()
        
    return median
    
# 4.
medians_dict = {key: get_median(medians_support_dict[key]) for key in medians_support_dict}

medians_dict
```

```python
# 5.
median_support_series = heart.age_dummy.astype(str) + '_' + heart.sex + '_' + heart.restingecg

median_support_series.head()
```

```python
# 6.
heart["cholesterol_medians"] = median_support_series.map(medians_dict)

heart["cholesterol_medians"].head()
```

```python
# 7.
index_cholest_zero = heart.query('cholesterol == 0').index

heart['cholesterol_2'] = heart['cholesterol']

heart.loc[index_cholest_zero, 'cholesterol_2'] = heart.loc[index_cholest_zero, 'cholesterol_medians']
```

As we can see below, this slice of `heart` shows that the invalid values in `cholesterol` were successfully replaced by values in `cholesterol_medians`. We can check it by comparing the medians in `cholesterol_2` with the medians in `medians_dict` via the keys.


```python
heart.query('cholesterol == 0')[['age', 'sex', 'restingecg', 'cholesterol', 'cholesterol_medians', 'cholesterol_2']].head()
```

### Heat map analysis of Pearson correlations.

```python
corr = heart.corr(method='pearson')

# Draw a heatmap with the numeric values in each cell.
f, ax = plt.subplots(figsize=(18, 12))

# Getting the Upper Triangle of the co-relation matrix.
matrix = np.triu(corr)

sns.heatmap(corr,
            annot=True,
            linewidths=.5,
            mask=matrix,
            ax=ax)

plt.show()
```

Given the results on the heat map, we can see that are almost no strong positive correlations between variables and strong negative correlation are rare as well. Focusing on the base of the triangle, which gives us the correlation of `heartdisease` with all the independent variables, we can see that there are no the strongest correlation is negative and regards a sub-category of the ST segment slope - `st_slope_up`. We'll include in the first version of the model the variables that have at least an absolute correlation of 0.3. Those are: 


```python
abs_corr_over_thres = corr.heartdisease.abs()[corr.heartdisease.abs() >= 0.3].sort_values(ascending=False)[1:]

abs_corr_over_thres
```

One variable that did not correlate strongly with `heardisease` although it was expected to, is `cholesterol_2`. Since as is, this feature is not correlated with the target variable `heartdisease` we'll check whether a dummy version of the same variable increases the correlation with the target. We we'll also trying a dummy version of the same variable that interacts with the former `age_dummy` variable to check if the target correlates with patients that are over 50 years old and have high cholesterol:
- `cholest_dummy`: `1` if (serum) cholesterol is equal or over 200 (these patients are considered unhealthy), else `0`.


- `high_cholest_over_50`: `1` if cholesterol is equal or over 200 and the patient is 50 years old or older, else `0`.

```python
heart["age_dummy"] = heart.age.apply(lambda x: 1 if x > 49 else 0)

heart['cholest_dummy'] = heart.cholesterol_2.apply(lambda x: 1 if x >= 200 else 0)

heart['high_cholest_over_50'] = heart.cholest_dummy * heart.age_dummy

heart['high_cholest_over_50'] = heart.cholest_dummy * heart.age_dummy

heart.corr()['heartdisease'][['cholest_dummy', 'high_cholest_over_50']]
```

As we can see the Pearson correlation values of both dummy variables are still under 0.3, thus we'll not include any version of the cholesterol-related features. Serum cholesterol is a broad measure of cholesterol that [includes varying portions of HDL cholesterol, LDL cholesterol and triglycerides](https://www.medicalnewstoday.com/articles/321519#what-is-it); unfortunately, we the information solely presented in the data set `heart` we cannot discriminate the levels of 'bad' cholesterol of each patient - LDL, from the other elements that compose serum cholesterol.

## Model Estimation
---
The data is to be split into train (60%), validation (20%) and test (20%) sets.

### Calculating accuracy scores for the validation set.

The function below, `obtain_score()`, takes training and testing sets, the features selected to serve as independent variables, the number of neighbors and the type of score to return: `'accuracy'`, `'sensitivity'` or `'specificity'`. The number of neighbors will be insert as part of a dictionary of `KNeighborsClassifier` parameters which include the weights attributed to each point and which frame is used to calculate the distances between points. For now we'll leave them at their default values: `weights='uniform'` and `p=2` (euclidean distance), changing only the number of neighbors.  

```python
def obtain_score(features, X_train, y_train, X_test, y_test, param_dict, score):
    
    knn = KNC(n_neighbors=param_dict['n'], weights=param_dict['weights'], p=param_dict['p'])
    
    length = y_test.size

    if isinstance(features, str):
        features = [features]
        
    X_train = X_train[features].values
    X_test = X_test[features].values
    y_train = y_train.values
    y_test = y_test.values
    
    model = knn.fit(X_train, y_train)
    
    prediction = model.predict(X_test)
    
    if score == 'accuracy':
        accuracy = sum(y_test == prediction) / length
        return round(accuracy*100, 2) # in percentage
    
    if score == 'sensitivity':
        true_pos = sum((y_test == 1) & (y_test == prediction))
        all_pos = y_test[y_test == 1].size # `all_pos` = true positives + false negatives
        sensitivity = true_pos / all_pos  
        return round(sensitivity*100, 2) # in percentage
    
    if score == 'specificity':
        true_neg = sum((y_test == 0) & (y_test == prediction))
        all_neg = y_test[y_test == 0].size # `all_neg` = true negatives + false positives
        specificity = true_neg / all_neg  
        return round(specificity*100, 2) # in percentage        
```

### Displaying the KNN classifier accuracy and sensitivity scores for univariate models.

In this experiment, we use the validation set to produce the predicted values and calculate the scores. To that end, we build a support function - `univar_df()`, that takes two random states which allow to randomly split `heart` into different train, validation and test set versions; it also takes an error score type (as a string) so that the function may return that calculated error score for each univariate model (features as columns) and for each model version (rows), where the number of neighbors `n` is set within a range between 1 and 10. 

```python
def univar_df(rand_stat_1, rand_stat_2, score):

    feats = [col for col in abs_corr_over_thres.index if col != 'heartdisease']

    X = heart[feats]

    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)


    univar_dict = {f: [] for f in feats}

    for f in univar_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            univar_dict[f].append(
                obtain_score(f, X_train, y_train, X_val, y_val, param_dict, score)
                )

    univar_df = pd.DataFrame(univar_dict, index=range(1, 10+1))

    return univar_df
```

The table above - `univar_accuracy_df` gives the accuracy results for univariate K-neighborhood Classifier estimations, for `n` neighbors ranging from 1 to 10. ## Multi-variable KNN classifier and error scores comparison
---
In this section we group all the selected variables filtered by the correlation analysis and estimate a multivariate KNN classifier, changing only the number of neighbors, in order to compare three different error scores: accuracy, sensitivity and specificity.

#### Normalizing  `oldpeak` in a min-max scale.

```python
scaler = MinMaxScaler()

heart['oldpeak_scaled'] =\
    scaler.fit_transform(np.array(heart.oldpeak).reshape(-1, 1)).reshape(heart.oldpeak.size).round(4)
```

#### Error scores comparison.

Similar to `univar_df()`, `multivar_df()` is a support function that returns a DataFrame that allows to compare error scores for the all-feature model, letting the number of neighbors vary from 1 to 10. The input is two random states that allow to created different shuffled versions of the train, validation and test sets. `mean_multivar_df` below, will gives the mean error scores of 100 different shuffles of the train and validation sets, which we'll use to calculate the scores.

```python
def multivar_df(rand_stat_1, rand_stat_2):


    feats = ['st_slope_up',
             'st_slope_flat',
             'exerciseangina_y',
             'oldpeak_scaled',
             'chestpaintype_ata',
             'maxhr',
             'sex_m']

    X = heart[feats]
    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)

    all_feats_dict = {'accuracy': [], 'sensitivity': [], 'specificity': []}

    for key in all_feats_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            all_feats_dict[key].append(
                obtain_score(feats, X_train, y_train, X_val, y_val, param_dict, key)
                )
            
    multivar_df = pd.DataFrame(all_feats_dict, index=range(1, 10+1))

    return multivar_df
```

### Grid search cross-validation of the all-feature model (hyperparameter tunning).

In this section we the Scikit-learn `GridSearchCV` frame in order to conduct a grid search that allows to retrieve error scores from several model estimations that combine different values for certain hyperparameters in `parameters_dict`:
- `n_neighbors`: number of neighbors raging from 1 to 20.
- `weights`: the weight attributed to each point - uniform or weighted.
- `p`: ["power parameter for the Minkowski metric"](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) ranging from 1 to 4.

The mean score values are calculated from a 5 folder cross-validation and the error scores calculated are accuracy and sensitivity (also designated as recall). `GridSearchCV` automatically divides the X and y input into train and validation sets, therefore, we split again `X_2` and `y_2` so that 80% is used to calculate the validation based error score while the remaining 20% will be used as the test set.  

```python
feats = ['st_slope_up',
         'st_slope_flat',
         'exerciseangina_y',
         'oldpeak_scaled',
         'chestpaintype_ata',
         'maxhr',
         'sex_m']

X = heart[feats]
y = heart.heartdisease

X_train, X_test, y_train, y_test = split(X, 
                                         y, 
                                         test_size=0.2, 
                                         random_state=100)

from sklearn.model_selection import GridSearchCV

parameters_dict = {
    'n_neighbors': list(range(1, 20+1)),
    'p': list(range(1, 4+1)),
    'weights': ['uniform', 'distance']
    }

knn = KNC()


grid_search_cv = GridSearchCV(estimator=knn,
                              param_grid=parameters_dict, 
                              scoring=['accuracy', 'recall'],
                              refit=False,
                              cv=5)


grid_cv = grid_search_cv.fit(X_train, y_train)

# Resume DataFrame - all hyperparameter combinations.
grid_cv_df = pd.DataFrame(grid_cv.cv_results_)

resume_cols = ['params',
               'mean_test_accuracy',
               'std_test_accuracy',
               'rank_test_accuracy',
               'mean_test_recall',
               'std_test_recall',
               'rank_test_recall']

grid_cv_df_short = grid_cv_df[resume_cols]
```
