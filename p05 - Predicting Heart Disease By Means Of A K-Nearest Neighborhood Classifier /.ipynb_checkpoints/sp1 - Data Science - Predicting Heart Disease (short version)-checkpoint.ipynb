# Predicting Heart Disease By Means Of A K-Nearest Neighborhood Classifier

## Introduction
---
Given a [data set](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction) with medical information on patients that have been subject to cardiovascular tests and profiling as well as a discrimination of those patients that have cardiovascular complications and those that don't. Our task is to design and train a k-nearest neighborhood (KNN) classifier that is able to predict successfully whether a patient has developed a heart complication or not, based on his/her medical profile.



|Feature | Description            |
|-----|---------------------------|
|Age  | age of the patient (years)|
|Sex  | sex of the patient (M: Male, F: Female)|
|ChestPainType  | chest pain type (TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: |Asymptomatic) |
|RestingBP  | resting blood pressure (mm Hg)|
|Cholesterol  | serum cholesterol (mm/dl)|
|FastingBS  | fasting blood sugar (1: if FastingBS > 120 mg/dl, 0: otherwise)|
|RestingECG  | resting electrocardiogram results (Normal: Normal, ST: having ST-T wave abnormality (T wave |inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria)|
|MaxHR  | maximum heart rate achieved (Numeric value between 60 and 202)|
|ExerciseAngina  | exercise-induced angina (Y: Yes, N: No)|
|Oldpeak  | oldpeak = ST (Numeric value measured in depression)|
|ST_Slope  | the slope of the peak exercise ST segment (Up: upsloping, Flat: flat, Down: downsloping)|
|HeartDisease  | output class (1: heart disease, 0: Normal)|

### Replacing invalid values in `cholesterol` with a sub-sample median.

Since `cholesterol` has 172 invalid observations we'll opt out by replacing those values with a median taken out of a sub-sample that controls for features that may affect the level of cholesterol, in this case: `age`, `sex` and `restingecg`. For `age` instead of using this variable as is, we are going to use a dummy version that simply distinguishes from patients under 50 years old and patients 50 years old and older. `sex` has also two subcategories and `restingecg` has three - `Normal`, `ST` and `LVH`, therefore the sub-samples we want to extract the medians from are combinations of those sub-categories. `medians_support_dict` contains all the combinations of these three features. The process to find the appropriate medians and replace the `cholesterol` invalid values with them is the following:


1. create the dummy variable for `age` in order to distinguish people with less than 50 years old from the rest.


2. create a dictionary - `medians_support_dict`, that will help to create sub-samples in `heart` with the combinations of `age_dummy`, `sex` and `restingecg`.


3. create a function - `get_median()`, that takes the values of `medians_support_dict`, which are lists with combinations of the sub-categories of the aforementioned features, fetches the correct sub-sample, and returns the associated median for that sub-sample. 


4. apply the function to the items in `medians_support_dict` in order to store the resulting medians in `medians_dict`. In both of these dictionaries each value/median is correctly associated with a 'code' string as key (same keys as in `medians_support_dict`) that are the combinations of the sub-categories of the aforementioned features: example `'50_and_over_M_Normal'` is a string that identifies the median for the sub-sample of observations that are 50 years old and older, male and have a normal resting ECG. 


5. create a support series - `median_support_series`, that we we'll use to map `medians_dict` in the next step. This series is build by adding the values of, in the following order, `age_dummy`, `sex` and `restingecg` (as strings), each element being separated by an underscore, "_".  


6. we apply the dictionary `medians_dict` in `median_support_series` to produce a series - `heart.cholesterol_medians` that maps correctly the sub-sample medians by their correct combination of features.  


7. inside the `heart` DataFrame, we clone `cholesterol` into a new column - `cholesterol_2`. We collect the index of the rows in `cholesterol` that are invalid (`0`), and use it to replace the same invalid values in `cholesterol_2` with the correspondent values (same index) in `heart.cholesterol_medians`.

```python
# 1.
heart["age_dummy"] = heart.age.apply(lambda x: "50_and_over" if x > 49 else "under_50")

# 2.
medians_support_dict = {
    '50_and_over_M_Normal': ['50_and_over', 'M', 'Normal'],
    '50_and_over_M_ST': ['50_and_over', 'M', 'ST'],
    '50_and_over_M_LVH': ['50_and_over', 'M', 'LVH'],
    'under_50_M_Normal': ['under_50', 'M', 'Normal'],
    'under_50_M_ST': ['under_50', 'M', 'ST'],
    'under_50_M_LVH': ['under_50', 'M', 'LVH'],
    '50_and_over_F_Normal': ['50_and_over', 'F', 'Normal'],
    '50_and_over_F_ST': ['50_and_over', 'F', 'ST'],
    '50_and_over_F_LVH': ['50_and_over', 'F', 'LVH'],
    'under_50_F_Normal': ['under_50', 'F', 'Normal'],
    'under_50_F_ST': ['under_50', 'F', 'ST'],
    'under_50_F_LVH': ['under_50', 'F', 'LVH'],
}

# 3.
def get_median(l):
    age = l[0]
    sex = l[1]
    ecg = l[2]
    
    median = heart.query("age_dummy == @age & sex == @sex & restingecg == @ecg")['cholesterol'].median()
        
    return median
    
# 4.
medians_dict = {key: get_median(medians_support_dict[key]) for key in medians_support_dict}

medians_dict
```

```python
# 5.
median_support_series = heart.age_dummy.astype(str) + '_' + heart.sex + '_' + heart.restingecg

median_support_series.head()
```

```python
# 6.
heart["cholesterol_medians"] = median_support_series.map(medians_dict)

heart["cholesterol_medians"].head()
```

```python
# 7.
index_cholest_zero = heart.query('cholesterol == 0').index

heart['cholesterol_2'] = heart['cholesterol']

heart.loc[index_cholest_zero, 'cholesterol_2'] = heart.loc[index_cholest_zero, 'cholesterol_medians']
```

As we can see below, this slice of `heart` shows that the invalid values in `cholesterol` were successfully replaced by values in `cholesterol_medians`. We can check it by comparing the medians in `cholesterol_2` with the medians in `medians_dict` via the keys.

```python
heart.query('cholesterol == 0')[['age', 'sex', 'restingecg', 'cholesterol', 'cholesterol_medians', 'cholesterol_2']].head()
```

## Model Estimation
---
The data is to be split into train (60%), validation (20%) and test (20%) sets.


### Calculating accuracy scores for the validation set.

The function below, `obtain_score()`, takes training and testing sets, the features selected to serve as independent variables, the number of neighbors and the type of score to return: `'accuracy'`, `'sensitivity'` or `'specificity'`. The number of neighbors will be insert as part of a dictionary of `KNeighborsClassifier` parameters which include the weights attributed to each point and which frame is used to calculate the distances between points. For now we'll leave them at their default values: `weights='uniform'` and `p=2` (euclidean distance), changing only the number of neighbors.  

```python
def obtain_score(features, X_train, y_train, X_test, y_test, param_dict, score):
    
    knn = KNC(n_neighbors=param_dict['n'], weights=param_dict['weights'], p=param_dict['p'])
    
    length = y_test.size

    if isinstance(features, str):
        features = [features]
        
    X_train = X_train[features].values
    X_test = X_test[features].values
    y_train = y_train.values
    y_test = y_test.values
    
    model = knn.fit(X_train, y_train)
    
    prediction = model.predict(X_test)
    
    if score == 'accuracy':
        accuracy = sum(y_test == prediction) / length
        return round(accuracy*100, 2) # in percentage
    
    if score == 'sensitivity':
        true_pos = sum((y_test == 1) & (y_test == prediction))
        all_pos = y_test[y_test == 1].size # `all_pos` = true positives + false negatives
        sensitivity = true_pos / all_pos  
        return round(sensitivity*100, 2) # in percentage
    
    if score == 'specificity':
        true_neg = sum((y_test == 0) & (y_test == prediction))
        all_neg = y_test[y_test == 0].size # `all_neg` = true negatives + false positives
        specificity = true_neg / all_neg  
        return round(specificity*100, 2) # in percentage        
```

### Displaying the KNN classifier accuracy and sensitivity scores for univariate models.

In this experiment, we use the validation set to produce the predicted values and calculate the scores. To that end, we build a support function - `univar_df()`, that takes two random states which allow to randomly split `heart` into different train, validation and test set versions; it also takes an error score type (as a string) so that the function may return that calculated error score for each univariate model (features as columns) and for each model version (rows), where the number of neighbors `n` is set within a range between 1 and 10. 

```python
def univar_df(rand_stat_1, rand_stat_2, score):

    feats = [col for col in abs_corr_over_thres.index if col != 'heartdisease']

    X = heart[feats]

    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)


    univar_dict = {f: [] for f in feats}

    for f in univar_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            univar_dict[f].append(
                obtain_score(f, X_train, y_train, X_val, y_val, param_dict, score)
                )

    univar_df = pd.DataFrame(univar_dict, index=range(1, 10+1))

    return univar_df
```

```python
univar_accuracy_df = univar_df(87, 323, 'accuracy')

univar_accuracy_df['best accuracy'] = univar_accuracy_df.max(axis=1) 

univar_accuracy_df.loc['idx_max', :] = univar_accuracy_df.idxmax(axis=0).astype('int32') 

univar_accuracy_df
```

The table above - `univar_accuracy_df` gives the accuracy results for univariate K-neighborhood Classifier estimations, for `n` neighbors ranging from 1 to 10. The last column - `best_accuracy`, identifies the best score along the same row (different independent variable but same `n`), and the last row - `idx_max`, identifies the model with the highest score when maintaining the independent variable but for different values of `n` (if there are repeated highest scores the `n` displayed is the smallest). The best model, when considering its accuracy, is the `st_slope_up` model, which arrives at the highest score when `n=5`: 78.14%. The second best model is `st_slope_flat` when `n=1`: 74.86%. `exerciseangina_y` appears in third place with a accuracy of 69.95% (`n=1`).

#### The sensitivity scores.

```python
univar_sensitivity_df = univar_df(87, 323, 'sensitivity')

univar_sensitivity_df['best sensitivity'] = univar_sensitivity_df.max(axis=1) 

univar_sensitivity_df.loc['idx_max', :] = univar_sensitivity_df.idxmax(axis=0).astype('int32') 

univar_sensitivity_df
```

For sensitivity, we observe a different scenario where `oldpeak` has the best prediction power - 94.17% (`n=5`), followed by `chestpaintype_ata` - 93.20% (`n=7`), `sex_m` - 92.23% (`n=1`) and then `st_slope_up` - 84.47% (`n=5`) and `st_slope_flat` - 75.73% (`n=1`). This values contrast a lot with we have seen with accuracy, which can be a symptom of fitting categorical variables while also, relying on a single pseudo-random distribution (random states) to shuffle the main data set into several smaller sets can; therefore, a more robust way to assess the sensitivity (and also the accuracy) is to calculate the mean of a large number of estimations, e.g. 100. So, lets add 99 versions of `univar_sensitivity_df` but with different random states and calculate the mean sensitivities.

```python
univar_mean_sensitivity_df = univar_df(87, 323, 'sensitivity')

for i in range(1, 100):

    univar_mean_sensitivity_df += univar_df(i, i+100, 'sensitivity')


univar_mean_sensitivity_df = univar_mean_sensitivity_df / 100

univar_mean_sensitivity_df['best sensitivity'] = univar_mean_sensitivity_df.max(axis=1) 

univar_mean_sensitivity_df.loc['idx_max', :] = univar_mean_sensitivity_df.idxmax(axis=0).astype('int32') 

univar_mean_sensitivity_df
```

As we can see in the table above - `mean_multivar_df`, the version that performs the best for `accuracy` and `sensitivity` is the model estimated with 3 neighbors: 79.06% and 83% respectively, whereas the model with the best `specificity` score is the one with 2 neighbors - 83.96%. Comparing the metrics for the model with 3 neighbors, we can see that its substantially better at predicting positive cases than negatives cases of heart disease (specificity at 74.81%). If we compare this results with the best univariate models, regarding mean accuracy and mean sensitivity, we arrive at the conclusion that the best univariate model in terms of mean accuracy - `st_slope_up`, has a slightly worst value - 78.80% (`n=9`), than the multivariate score. Similarly, regarding sensitivity, the multivariate version has a slightly better mean sensitivity score (83%), than the `st_slope_up` univariate model - 81.25%. What we can conclude from this comparison is that due to either the current design of the model or the features included, all the features besides `st_slope_up` do not contribute to a substantially better prediction of the target variable.

```python
univar_mean_accuracy_df = univar_df(87, 323, 'accuracy')

for i in range(1, 100):

    univar_mean_accuracy_df += univar_df(i, i+100, 'accuracy')


univar_mean_accuracy_df = univar_mean_accuracy_df / 100

univar_mean_accuracy_df['best accuracy'] = univar_mean_accuracy_df.max(axis=1) 

univar_mean_accuracy_df.loc['idx_max', :] = univar_mean_accuracy_df.idxmax(axis=0).astype('int32') 

univar_mean_accuracy_df
```

Compared with the previous accuracy results shown in `univar_accuracy_df`, we can assess that the general landscape is about the same, where the highest scores in `univar_mean_accuracy_df` can be found for `st_slope_up`, which arrives at a maximum accuracy of 78.80% at `n=9`, being a slightly higher value than the analogous accuracy score observed in `univar_accuracy_df` (78.14%); and for `st_slope_flat`, which is still the second best accuracy predictor, and its highest value also slightly increases from 74.86%(`n=1`, in `univar_accuracy_df`) to current value of 76.42% (`n=10`). `exerciseangina_y` appears in third place, as before, maintaining a accuracy around 70%


## Multi-variable KNN classifier and error scores comparison
---
In this section we group all the selected variables filtered by the correlation analysis and estimate a multivariate KNN classifier, changing only the number of neighbors, in order to compare three different error scores: accuracy, sensitivity and specificity.

#### Normalizing  `oldpeak` in a min-max scale.

```python
scaler = MinMaxScaler()

heart['oldpeak_scaled'] =\
    scaler.fit_transform(np.array(heart.oldpeak).reshape(-1, 1)).reshape(heart.oldpeak.size).round(4)
```

#### Error scores comparison.

Similar to `univar_df()`, `multivar_df()` is a support function that returns a DataFrame that allows to compare error scores for the all-feature model, letting the number of neighbors vary from 1 to 10. The input is two random states that allow to created different shuffled versions of the train, validation and test sets. `mean_multivar_df` below, will gives the mean error scores of 100 different shuffles of the train and validation sets, which we'll use to calculate the scores.

```python
def multivar_df(rand_stat_1, rand_stat_2):


    feats = ['st_slope_up',
             'st_slope_flat',
             'exerciseangina_y',
             'oldpeak_scaled',
             'chestpaintype_ata',
             'maxhr',
             'sex_m']

    X = heart[feats]
    y = heart.heartdisease

    # The `test_size` here refers to the `rest` sets.
    X_train,  X_rest, y_train, y_rest = split(X,
                                              y,
                                              test_size=0.4,
                                              random_state=rand_stat_1)

    X_val, X_test, y_val, y_test = split(X_rest,
                                         y_rest,
                                         test_size=0.5,
                                         random_state=rand_stat_2)

    all_feats_dict = {'accuracy': [], 'sensitivity': [], 'specificity': []}

    for key in all_feats_dict:
        for i in range(1, 10+1):
            param_dict = {'n': i, 'weights': 'uniform', 'p': 2}
            all_feats_dict[key].append(
                obtain_score(feats, X_train, y_train, X_val, y_val, param_dict, key)
                )
            
    multivar_df = pd.DataFrame(all_feats_dict, index=range(1, 10+1))

    return multivar_df
```

```python
sum_multivar_df = None

for i in  np.arange(0, 200, 2):
    if sum_multivar_df is None:
        sum_multivar_df  = multivar_df(i, i+1)
    else: 
        sum_multivar_df  += multivar_df(i, i+1)

mean_multivar_df = sum_multivar_df / 100

mean_multivar_df.loc['idx_max', :] = mean_multivar_df.idxmax(axis=0).astype('int32') 
```

```python
mean_multivar_df
```

As we can see in the table above - `mean_multivar_df`, the version that performs the best for `accuracy` and `sensitivity` is the model estimated with 3 neighbors: 79.41% and 83.21% respectively, whereas the model with the best `specificity` score is the one with 2 neighbors - 83.81%. Comparing the metrics for the model with 3 neighbors, we can see that its substantially better at predicting positive cases than negatives cases of heart disease (specificity at 74.81%). If we compare this results with the best univariate models, regarding mean accuracy and mean sensitivity, we arrive at the conclusion that the best univariate model  in terms of mean accuracy - `st_slope_up`, has a slightly better value - 80.73% (`n=7`), than the multivariate score. The same happens for sensitivity: the `st_slope_up` univariate model has a slightly better mean sensitivity score - 83.75% (`n=7`), than multivariate version - 83.21%. This is important because it tells us that the due to either the current design of the model or the features included, all the features besides `st_slope_up` do not contribute to a better prediction of the target variable.

### Grid search cross-validation of the all-feature model (hyperparameter tunning).

In this section we the Scikit-learn `GridSearchCV` frame in order to conduct a grid search that allows to retrieve error scores from several model estimations that combine different values for certain hyperparameters in `parameters_dict`:
- `n_neighbors`: number of neighbors raging from 1 to 20.
- `weights`: the weight attributed to each point - uniform or weighted.
- `p`: ["power parameter for the Minkowski metric"](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) ranging from 1 to 4.

The mean score values are calculated from a 5 folder cross-validation and the error scores calculated are accuracy and sensitivity (also designated as recall). `GridSearchCV` automatically divides the X and y input into train and validation sets, therefore, we split again `X_2` and `y_2` so that 80% is used to calculate the validation based error score while the remaining 20% will be used as the test set.  

```python
feats = ['st_slope_up',
         'st_slope_flat',
         'exerciseangina_y',
         'oldpeak_scaled',
         'chestpaintype_ata',
         'maxhr',
         'sex_m']

X = heart[feats]
y = heart.heartdisease

X_train, X_test, y_train, y_test = split(X, 
                                         y, 
                                         test_size=0.2, 
                                         random_state=100)

from sklearn.model_selection import GridSearchCV

parameters_dict = {
    'n_neighbors': list(range(1, 20+1)),
    'p': list(range(1, 4+1)),
    'weights': ['uniform', 'distance']
    }

knn = KNC()


grid_search_cv = GridSearchCV(estimator=knn,
                              param_grid=parameters_dict, 
                              scoring=['accuracy', 'recall'],
                              refit=False,
                              cv=5)


grid_cv = grid_search_cv.fit(X_train, y_train)

# Resume DataFrame - all hyperparameter combinations.
grid_cv_df = pd.DataFrame(grid_cv.cv_results_)

resume_cols = ['params',
               'mean_test_accuracy',
               'std_test_accuracy',
               'rank_test_accuracy',
               'mean_test_recall',
               'std_test_recall',
               'rank_test_recall']

grid_cv_df_short = grid_cv_df[resume_cols]
```

#### The best three models according to mean accuracy and mean sensitivity (recall):

```python
highest_accuracy_hyper = grid_cv_df_short.sort_values('mean_test_accuracy', ascending=False).head(3)

highest_accuracy_hyper
```

In terms of accuracy we can see that the best three models, displayed in `highest_accuracy_hyper`, have uniform weights and the same distance power parameter - `p=1`, (Manhattan distance); we can also see that despite of the change in the number of neighbors, `n`, the score value is in all cases roughly 80%.


```python
highest_sensitivity_hyper = grid_cv_df_short.sort_values('mean_test_recall', ascending=False).head(3)

highest_sensitivity_hyper
```

In terms of sensitivity the weights and distance are also uniform and Manhattan respectively, as previously, but the model with highest mean sensitivity is achieved when `n=11`: 80.11%. Again the number of neighbors doesn't seem to affect to provide a substantially better score than the rest.

## Conclusion
---
Resuming the findings, the univariate KNN classifier that only includes a dummy variable that distinguishes patients that show a up sloping ST segment in their electrocardiogram, rather than a flat or down sloping behavior - `st_slope_up`, was as good as a multivariate version of the classifier in predicting heart disease, with an accuracy and sensitivity around the 80% score mark. In opposition, we could not increase the prediction power of the model by including variables that showed a correlation of 30% and higher with the target variable place. These variables were:

|Feature             |Description                                               |
|--------------------|----------------------------------------------------------|
|'st_slope_flat'     | '1' if ST segment is flat, else '0'                      |
|'exerciseangina_y'  | '1' if exercise-induced angina, else '0'                 |
|'oldpeak_scaled'    | ST depression induced by exercise relative to rest (mm). |
|'chestpaintype_ata' | '1' if chest pain is derived from atypical angina (ata), else '0' |
|'maxhr'             |  maximum heart rate achieved                             |
|'sex_m'             |  '1' if male, 0 if female                                |

In terms of feature inclusion into the model, we opted out to exclude (serum) `cholesterol`, even though we tried to modify this variable by distinguish patients with more than 50 years and cholesterol over 200 mm/dl, in order to attempt to capture a correlation with the target variable over 30%, which we were not able to establish. 

We could also found that optimizing the hyperparameters according to a search grid corss-validation process - `n` (the number of neighbors), `wheights` (the weight attributed to each point in a neighborhood) and `p` (the power parameter of the minkowski distance used to calculate distances between neighbors), did not produce improved prediction scores for both the univariate (`st_slope_up`) and multivariate model, maintaining the accuracy and sensitivity scores around the 80% mark. 

Since we could find a univariate model that is able to predict the outcome up to 80%, there is space for improvement; this can be achieved through different methods such as including other important features to the analysis that were not present in the 'heart_disease_prediction.csv' data set; examples of those features are the level of LDL cholesterol per patient; a categorical variable that describes the smoking behavior of the patient e.g. 'non-smoker', 'light-smoker' and 'heavy smoker'; or even the body mass index (BMI) of the patient. Other methods could entail to better engineer the current features, resorting to a deeper understanding of the field of cardiology, or even to try to apply other methodologies used for the purpose of prediction e.g. decision tree classifier, random forest classifier, logistic regression.
